```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&ackey=c6s6adu6"

resp = requests.get(url)
resp # 정상적인 응답 확인
```

```python
soup = BeautifulSoup(resp.text, 'lxml')
soup
```

```python
# 위키피디아 미국 ETF 웹 페이지에서 필요한 정보를 스크래핑하여 딕셔너리 형태로 변수 etfs에 저장
url = "https://en.wikipedia.org/wiki/List_of_American_exchange-traded_funds"
resp = requests.get(url)
resp
```

```python
soup = BeautifulSoup(resp.text, 'lxml')   
soup
```

```python
rows = soup.select('div > ul > li')
rows
```

```python
print("resp: ", resp)
print("=" * 100)
print("resp.text: ", resp.text)
print("=" * 100)
print("soup: ", soup)
print("=" * 100)
print("rows: ", rows[50:55])
```

```python
rows[50:55]
```

```python
etfs = {}
for row in rows:
    
    try:
        etf_name = re.findall("(.+?)\s\(", row.text) #re모듈은 정규표현식으로 텍스트를 선별적으로 추출
        print(etf_name)

    except AttributeError as err:
        pass    

# etfs 딕셔너리 출력
print(etfs)
```

```python
etfs = {}
for row in rows:
    
    try:
        etf_market = re.findall("\((.+?):", row.text) #(.+?) - 모든문자(최소탐욕) -가능한한 짧게 매칭
        print(etf_market)

    except AttributeError as err:
        pass    

# etfs 딕셔너리 출력
print(etfs)
```

```python
etfs = {}
for row in rows:
    
    try:
         etf_ticker = re.findall(":\s(.+)\)", row.text) # \s 공백문자 하나, \(여는 괄호)
         print(etf_ticker)
    
    except AttributeError as err:
        pass    

# etfs 딕셔너리 출력
print(etfs)
```

```python
etfs = {}
for row in rows:
    
    try:
        etf_name = re.findall("(.+?)\s\(", row.text) #re모듈은 정규표현식으로 텍스트를 선별적으로 추출
        etf_market = re.findall("\((.+?):", row.text) #(.+?) - 모든문자(최소탐욕) -가능한한 짧게 매칭
        etf_ticker = re.findall(":\s(.+)\)", row.text) # \s 공백문자 하나, \(여는 괄호)
        
        if (len(etf_ticker) > 0) & (len(etf_market) > 0) & (len(etf_name) > 0):
            etfs[etf_ticker[0]] = [etf_market[0], etf_name[0]]

    except AttributeError as err:
        pass    

# etfs 딕셔너리 출력
etfs
```

```python
df = pd.DataFrame(etfs)
df
```
